#### Get Nasdaq + SP500 + Dow + Russell 3000 Data from YFinance
* Tickers
  - '^SPX': 'S&P 500'
  - '^IXIC': 'Nasdaq'
  - '^RUA': 'Russell 3000'
  - '^DJI': 'Dow'


````python
import yfinance as yf
import pandas as pd
import plotly.express as px
symbol_to_name_map = {'^SPX': 'S&P 500', '^IXIC': 'Nasdaq', '^RUA': 'Russell 3000', '^DJI': 'Dow'}
symbol_list = list(symbol_to_name_map.keys())

data = yf.download(tickers=symbol_list,period='max',interval='1mo')

data
````

Take just adjusted close

````python
data = data['Adj Close']

data
````

#### Calculate Monthly Returns

````python
returns = data.pct_change()

returns
````

### Get Month Name

````python
returns['Month']=returns.index.month_name()
returns
````

**Statistical Test: One-Sided Independent T-Test for means**

This test is used to compare the means of two independent samples (September returns and returns from other months) and determine if there's a statistically significant difference, specifically if one mean is significantly lower than the other.

**Equations**

1. **T-statistic:**
   
   ```
   t = (x̄₁ - x̄₂) / √((s₁²/n₁) + (s₂²/n₂))
   ```

   where:
   * x̄₁: Mean of September returns
   * x̄₂: Mean of returns from other months
   * s₁: Standard deviation of September returns
   * s₂: Standard deviation of returns from other months
   * n₁: Number of observations in September returns
   * n₂: Number of observations in returns from other months

2. **Degrees of Freedom:**

   The degrees of freedom for the Welch's t-test (assuming unequal variances) is calculated using the Welch-Satterthwaite equation:

   ```
   df ≈ ((s₁²/n₁ + s₂²/n₂)²)/(((s₁²/n₁)²/(n₁-1)) + ((s₂²/n₂)²/(n₂-1)))
   ```

3. **One-sided P-value:**

   ```
   p_value_one_sided = p_value / 2
   ```

   where:
   * p_value: Two-sided p-value obtained from the t-test

**Bullet Points - Test Procedure**

*   **Hypotheses:**
    *   Null Hypothesis (H0): The mean return for September is equal to or greater than the mean return for other months.
    *   Alternative Hypothesis (H1): The mean return for September is less than the mean return for other months.

*   **Calculate T-statistic:** Compute the t-statistic using the means, standard deviations, and sample sizes of the two groups.

*   **Calculate Degrees of Freedom:** Determine the degrees of freedom using the Welch-Satterthwaite equation.

*   **Obtain P-value:** Use the t-statistic and degrees of freedom to find the two-sided p-value from the t-distribution.

*   **Calculate One-sided P-value:** Divide the two-sided p-value by 2 to get the one-sided p-value for the lower tail.

*   **Decision:**
    *   If the one-sided p-value is less than the chosen significance level (alpha) AND the t-statistic is negative, reject the null hypothesis. This indicates that September returns are significantly lower than other months' returns.
    *   Otherwise, fail to reject the null hypothesis. There's not enough evidence to conclude that September returns are significantly worse.

**Remember:**

*   We assume unequal variances between the two groups in this test.
*   Outlier handling is crucial before performing the test to ensure the robustness of the results.
*   Interpretation should be done in context, considering other factors that might influence returns besides the month.




````python
from scipy.stats import ttest_ind
from scipy.stats import ttest_1samp, zscore


# ... (your existing data loading and preprocessing code) ...
for symbol in symbol_list:
  symbol_name = symbol_to_name_map[symbol]
  # 1. Separate September and other months returns
  september_returns = returns[symbol][returns['Month'] == 'September']
  other_months_returns = returns[symbol][returns['Month'] != 'September']

  # 2. Drop missing values (NaN) as t-test requires numeric data
  september_returns.dropna(inplace=True)
  other_months_returns.dropna(inplace=True)

  # 3. Perform the t-test
  t_statistic, p_value = ttest_1samp(september_returns, other_months_returns.mean())

  # 4. Interpret results
  alpha = 0.05
  if p_value < alpha:
      print(f"Reject null hypothesis. September returns are statistically significantly worse for {symbol_name} at {alpha} significance.")
  else:
      print("Fail to reject null hypothesis. No significant difference in returns.")

  # Print additional information
  print(f"T-statistic: {t_statistic}")
  print(f"P-value: {p_value}")
````

### What about other months?

````python
list_of_months = returns.Month.unique().tolist()
````

````python
from scipy.stats import ttest_ind
from scipy.stats import ttest_1samp, zscore

for symbol in symbol_list:
  symbol_name = symbol_to_name_map[symbol]
  for month in list_of_months:
    if month == 'September': continue
    # 1. Separate September and other months returns
    month_returns = returns[symbol][returns['Month'] == month]
    other_months_returns = returns[symbol][returns['Month'] != month]

    # 2. Drop missing values (NaN) as t-test requires numeric data
    month_returns.dropna(inplace=True)
    other_months_returns.dropna(inplace=True)

    # 3. Perform one-sided t-test
    t_statistic, p_value = ttest_ind(month_returns, other_months_returns, equal_var=False)

    # Calculate one-sided p-value
    p_value_one_sided = p_value / 2  # Divide two-sided p-value by 2 for lower tail test

    # 4. Interpret results (adjusted for one-sided test)
    alpha = 0.05
    if p_value_one_sided < alpha and t_statistic < 0:  # Check if t-statistic is negative (lower tail)
      print(f"Reject null hypothesis. {month} returns are statistically significantly worse for {symbol_name} at {alpha} significance.")
    else:
        print(f"Fail to reject null hypothesis. No significant evidence that {month} returns are worse.")
````
