#### Get Russell 1000 stocks with *pd.read_html*




```python
import pandas as pd
df_html = pd.read_html('https://en.wikipedia.org/wiki/Russell_1000')
```

* Inspect the tables

```python
from IPython.display import display,Markdown
for i,table in enumerate(df_html):
  display(Markdown(f'###Table {i}'))
  display(table)
```

* Get the stock symbols



```python
symbol_list = df_html[2].Symbol.tolist()
len(symbol_list)
```

* Someone commented about not seeing TSM (well I guess it isn't part of Russell 1000)

```python
'TSM' in symbol_list
```

#### 2. Get Company Info from Yahoo Finance API

* Market Cap, Industry,Sector



```python
from IPython.display import display,Markdown
import yfinance as yf

info_list = []

for i,symbol in enumerate(symbol_list):
  if i%100==0:
    print(f"Completed {i}/{len(symbol_list)}")
  try:
    info = yf.Ticker(symbol).info
    info['symbol'] = symbol
    info_list.append(info)
  except:
    pass
```

* Turn into a dataframe

```python
df_fundamental = pd.DataFrame(info_list)
df_fundamental.head()
```

 Look at Market *cap* to see the numbers make sense

```python
df_fundamental[['symbol','marketCap']].sort_values('marketCap',ascending=False)
```

```python
import plotly.express as px
fig = px.bar(df_fundamental[['symbol','marketCap']].nlargest(200,'marketCap').sort_values('marketCap',ascending=True),x='marketCap',y='symbol',height=3000)
fig.show()
```

* AVGO seems off by a factor of 10! so let's divide by 10!

```python
df_fundamental['marketCap']=df_fundamental.apply(lambda l:l['marketCap']/10 if l['symbol']=='AVGO' else l['marketCap'],axis=1)
```

#### 3. Get actual OHLCV for 1000+ stock symbols at once
* We only need Thursday 2024-07-11's returns so need data from 2024-07-10


```python
data = yf.download(symbol_list,start='2024-07-10',end='2024-07-12')
```

* Inspect the results

```python
data.head()
```

* Calculate pct_change using only adjusted close which adjust for dividends



```python
df_pct_change=data['Adj Close'].pct_change().iloc[1:,:].T.reset_index()
df_pct_change
```

* Clean up the columns

```python
df_pct_change.columns=['symbol','pct_change']
df_pct_change.head()
```

#### Merge with fundamental data and drop missing values
* We only will be using symbol,sector,marketCap, industry,and pct_change


```python
df = df_pct_change.merge(df_fundamental[['symbol','sector','marketCap','industry']],on='symbol').dropna(axis=0)
df.head()
```

#### 4. Fun Colorful Part: Plotly TreeMap Vizualization
* It partitions the heatmap into smaller and smaller subset using the "path" argument
  - If you want an intial subset for everything you can specify it with px.Constant
* You can specify a custom colorscale.
  - Red to green depending on stock performance
  - White as the neutral 0% change point
  
* https://plotly.com/python/treemaps/

```python
fig = px.treemap(
    data_frame=df,
    path = [px.Constant("Russell 1000"),"sector","industry","symbol"],
    values = 'marketCap',hover_data=['symbol','sector','industry','marketCap']

)
fig.show()
```

```python
color_scale = [
    [0,'darkred'],
    [0.4,'red'],
    [0.45,'salmon'],
    [0.5,'white'],
    [0.55,'lightgreen'],
    [0.6,'green'],
    [1,'darkgreen']
]

fig = px.treemap(
    data_frame=df,
    path = [px.Constant("Russell 1000"),"sector","industry","symbol"],
    values = 'marketCap',hover_data=['symbol','sector','industry','marketCap'],
    color = 'pct_change',color_continuous_scale=color_scale,color_continuous_midpoint=0
)
fig.show()
```
