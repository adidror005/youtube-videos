### Connect to IB

```python
from ib_insync import *
util.startLoop()
ib = IB()
ib.connect(port=4002,clientId=1)
```

### MetaData

```python
metadata = ib.getWshMetaData()
metadata
```

```python
import json
metadata = json.loads(metadata)
metadata
```

```python
metadata['meta_data']['filters']
                      
```

```python
event_types_df=util.df(metadata['meta_data']['event_types'])
event_types_df                   
```

## Get actual event data

```python
symbol_list = ['TSLA','AAPL','META','AMZN','RIVN']
stocks = [Stock(symbol=symbol,exchange='SMART',currency='USD') for symbol in symbol_list]
ib.qualifyContracts(*stocks)
conid_list = [str(s.conId) for s in stocks]
conid_list
```

```python
filter_string = json.dumps({
    "watchlist": conid_list
})
data = WshEventData(filter=filter_string)
data
```

```python
events = ib.getWshEventData(data)
events
```

```python
events = json.loads(events)
events
```

```python
df_events = util.df(events)
df_events
```

```python
df_events = df_events.merge(event_types_df,left_on='event_type',right_on='tag')
```

```python
df_events['name'].value_counts()
```

```python
df_events['data'].tolist()
```
